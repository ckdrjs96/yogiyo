```python
document_embedding_list = []

for review in tokenized_data:
    doc2vec = None
    cnt=0
    for word in review:
        if word in pre_model.wv.vocab:
            cnt+=1
            if doc2vec is None:
                doc2vec= pre_model.wv.__getitem__(word)
            else:
                doc2vec= doc2vec+ pre_model.wv.__getitem__(word)
        
    if doc2vec is not None:
            
        doc2vec=doc2vec/cnt
        document_embedding_list.append(doc2vec)
    
    else:
        print(review)
    
            
        
```

```python
cosine_similarities = cosine_similarity(document_embedding_list, document_embedding_list)
```
