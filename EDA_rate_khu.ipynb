```python
def condition_3(tot_star,flavor,volume,deliver,imgname):
    imgname= ast.literal_eval(imgname)
    if len(imgname) != 0 and tot_star==5 and flavor ==5 and volume==5 and deliver==5:
        return True
    else:
        return False
```

```python
def condition_2(tot_star,flavor,volume,deliver,imgname):
    if tot_star==5 and flavor ==5 and volume==5 and deliver==5:
        return True
    else:
        return False
```

```python
def condition_1(tot_star,flavor,volume,deliver,imgname):
    imgname= ast.literal_eval(imgname)
    if len(imgname) != 0:
        return True
    else:
        return False
```

```python
univ='khu'
path=f'/data/yogiyo/pr_data/{univ}/'
shops1=pd.read_csv(path+f'{univ}_shop.csv')
```

```python
cata4_1=pd.read_csv(path+'cata4_1.csv')
cata4_2=pd.read_csv(path+'cata4_2.csv')
cata4_3=pd.read_csv(path+'cata4_3.csv')
```

```python
cata4=pd.concat([cata4_1,cata4_2,cata4_3])
```

```python
cata0=pd.read_csv(path+'cata0.csv')
cata1=pd.read_csv(path+'cata1.csv')
cata2=pd.read_csv(path+'cata2.csv')
cata3=pd.read_csv(path+'cata3.csv')
```

```python
def what_condition(i,fuc):
    cata0[f'condition{i}']=cata0.apply(lambda x: fuc(x.tot_star,x.star_flavor,x.star_volume,x.star_deliver,x.img_name), axis=1)
    cata1[f'condition{i}']=cata1.apply(lambda x: fuc(x.tot_star,x.star_flavor,x.star_volume,x.star_deliver,x.img_name), axis=1)
    cata2[f'condition{i}']=cata2.apply(lambda x: fuc(x.tot_star,x.star_flavor,x.star_volume,x.star_deliver,x.img_name), axis=1)
    cata3[f'condition{i}']=cata3.apply(lambda x: fuc(x.tot_star,x.star_flavor,x.star_volume,x.star_deliver,x.img_name), axis=1)
    cata4[f'condition{i}']=cata4.apply(lambda x: fuc(x.tot_star,x.star_flavor,x.star_volume,x.star_deliver,x.img_name), axis=1)
```

```python
what_condition(1,condition_1)
what_condition(2,condition_2)
what_condition(3,condition_3)
```

## 카테고리조건 일치/불일치에 따른 별점비교

catagory1조건(사진을 요구한다) 일치/불일치의 별점비교

```python
pd.set_option('display.max_rows',None)
a=pd.pivot_table(cata1,index='shop',columns='condition1',values='tot_star',aggfunc=['mean','count'])
b=shops.iloc[:,:3]
c=pd.merge(a,b,left_on='shop',right_on='shop',how='left').sort_values(by='star', ascending=False)
c[c.review_cnt >20]
```

catagory2조건(별5개를요구한다) 일치/불일치의 별점비교

```python
a=pd.pivot_table(cata2,index='shop',columns='condition2',values='tot_star',aggfunc=['mean','count'])
b=shops.iloc[:,:3]
c=pd.merge(a,b,left_on='shop',right_on='shop',how='left').sort_values(by='star', ascending=False)
c[c.review_cnt >20]
```

catagory3조건(사진/별5개를요구한다) 일치/불일치의 별점비교

```python
a=pd.pivot_table(cata3,index='shop',columns='condition3',values='tot_star',aggfunc=['mean','count'])
b=shops.iloc[:,:3]
c=pd.merge(a,b,left_on='shop',right_on='shop',how='left').sort_values(by='star', ascending=False)
c[c.review_cnt >20]
```

catagory4(리뷰이벤트를 하지 않음)을 조건3(사진/별5개를요구한다)으로 했을떄 일치/불일치의 별점비교 

```python
a=pd.pivot_table(cata4,index='shop',columns='condition3',values='tot_star',aggfunc=['mean','count'])
b=shops.iloc[:,:3]
c=pd.merge(a,b,left_on='shop',right_on='shop',how='left').sort_values(by='star', ascending=False)
c[c.review_cnt >20]
```
