```python
import os, re
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from konlpy.tag import Okt
from collections import Counter
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
```

```python
PATH='/root/workspace/simdata/'
```

```python
cat15=pd.read_csv(PATH+'cat15_sim.csv')

cat11=pd.read_csv(PATH+'cat11_sim.csv')

cat7=pd.read_csv(PATH+'cat7_sim.csv')

cat3=pd.read_csv(PATH+'cat3_sim.csv')
```

```python
cat3=pd.concat([cat15,cat11,cat7,cat3])
```

```python
cat3=cat3.drop(['token','sim','sim2'],axis=1)
```

```python
cat3
```

```python
def isimg(imgname):
    if imgname == imgname:
        return True
    else:
        return False
```

```python
cat3['isimg']=cat3.img_name.map(lambda x: isimg(x))

cat3['isstarflavor']=cat3.star_flavor.map(lambda x: True if x==5 else False)

cat3['isstarvolume']=cat3.star_volume.map(lambda x: True if x==5 else False)

cat3['isstardeliver']=cat3.star_deliver.map(lambda x: True if x==5 else False)

cat3['final']=cat3.apply(lambda x: True if x.isimg & x.isstarflavor & x.isstarvolume & x.isstardeliver==True else False, axis=1)
```

```python
cat3.final.value_counts()
```
