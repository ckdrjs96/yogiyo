```python
def find_in(x):
    find=['리뷰작성','리뷰 작성','리뷰 이벤트','리뷰이벤트','후기이벤트','후기 이벤트','리뷰','서비스','써비스','참여','증정','약속']
    cnt=0
    for tok in find:
        if re.search(tok, x):
            cnt+=1
        if cnt ==2:
            return True
    return False
```

```python
isservice=china_isnotice['notice'].map(lambda x: find_in(x))
china_isnotice['isservice']=isservice
```

```python
chicken_notice=pd.read_csv(PATH+'chicken_notice.csv', encoding='utf-8')
```

```python
def preprocess(notice):
    del notice['Unnamed: 0']
    notice=notice.dropna().reset_index(drop=True)
    isservice=notice['notice'].map(lambda x: find_in(x))
    notice['isservice']=isservice
    
    return notice
```

```python
def check_num(notice,isnotice):
    print(notice.shape)
    print(isnotice.shape)
    print(sum(isnotice.isservice))
    
```
