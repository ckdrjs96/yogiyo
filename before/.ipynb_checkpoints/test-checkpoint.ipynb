```python
def find_in(x):
    find=['리뷰','이벤트','서비스','써비스','참여','증정']
    cnt=0
    for tok in find:
        if re.search(tok, x):
            cnt+=1
        if cnt ==2:
            return True
    return False
```

```python
isservice=china_isnotice['notice'].map(lambda x: find_in(x))
```
